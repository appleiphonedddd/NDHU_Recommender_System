```python
columns_names=['user_id','item_id','rating','timestamp']
df=pd.read_csv("u.data", sep="\t",names=columns_names)
df.head(2)
```

#  Let's Import Second Data Set

```python
columns_names=['item_id','title','date', "A1", "A2", "A3", "A4", "A5", "A6", "A7", "A8", "A9", "A10", "A11", "A12", "A13", "A14", "A15", "A16", "A17", "A18", "A19", "A20", "A21"]

movies = pd.read_csv('u.item', sep='|', encoding='latin-1', names=columns_names)
movies.head(2)
```

 We need just first two colums

```python
movies=movies.iloc[: , 0:2]
movies.head()
```

```python
movies.shape
```

# Merge

```python
df = pd.merge(df, movies, on='item_id')
df.head()
```

We won't use the timestamp, let's drop it;

# Drop

```python
df.drop(["timestamp"], axis = 1, inplace = True)

df.head(2)
```

# Let's Set Up Our Recommendation System


+ Firstly, we set up a structure similar to the pivot table in Excel.
+ According to this structure, each row will be a user (so the index of our data frame will be user_id)
+ The columns will have film names (i.e. the title column),
+ We create a data frame with rating values in the table!

```python
moviemat = df.pivot_table(index='user_id',columns='title',values='rating')
moviemat.head()
```

```python
type(moviemat)
```

# Aim: To offer users with recommendations for films similar to Starwars

We will use the **Item - Item recommodition** method, selecting and recommending similar films (those with similar ratings).  

We will make a **collaborative recommodition** using the ratings given by the users.

We will recommend the **most liked** (leader) film to everyone.

```python
# Let's look at the user ratings of the film Star Wars (1977):

starwars_user_ratings = moviemat['Star Wars (1977)']
starwars_user_ratings.head()
```

# corrwith()

Let's calculate the correlation of all films with the Star Wars film using the corrwith() method:

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
similar_to_starwars

# It gives the following error due to NaN values.
```

```python
similar_to_starwars
```

+ According to this result, the correlation between the Star Wars film and the film "Til There Was You (1997)" is 0.87. In other words, 87% of the people who watched the Star Wars film also watched the film "Til There Was You (1997)".

+ On the other hand, the correlation between the film "Tar Wars" and the film "1-900 (1994)" was -0.64. In other words, 64% of the people who watched the Star Wars film did not watch the film "1-900 (1994)".

```python
type(similar_to_starwars)
```

* The list we've got is a Pandas series. If we convert it to df, we can work more easily.

```python
corr_starwars = pd.DataFrame(similar_to_starwars, columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

# Let's list the data frame we obtained and see which film it suggests as the closest to Star Wars.

```python
corr_starwars.sort_values('Correlation', ascending=False).head(10)
```

+ As you can see, there were irrelevant results, when you research this subject a little, you will find that the reason for this is that these films received very few votes. Or; if someone who gave 5 to Star Wars also gave 5 to "Hollow Reed (1996)", the correlation is 1, but if only a few people gave 5 to the second film, the correlation result obtained may be misleading.

+ To fix this, let's eliminate films with less than 100 votes. For this purpose, let's create a dataframe called ratings and keep track of how many votes each film has received (i.e. the number of votes).

```python
df.head()
```

# Let's find the average (mean value) rating of each film

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings
```

# Let's sort it

```python
ratings.sort_values('rating', ascending=False).head(20)
```

**Attention:** When calculating these averages, we didn't look at how many votes they received, so there are films like this that we don't know at all. Probably there are also films that were voted only once. This makes us wrong.

# Now let's find the number of votes each film received

```python
ratings['rating_vote_number'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

# Now let's take a look at the most voted films in descending order

```python
ratings.sort_values('rating_vote_number',ascending=False).head()
```

# Let's go back to our main purpose and add the rating_vote_number column to our corr_starwars dataframe (with join)

```python
corr_starwars.sort_values('Correlation', ascending=False).head(10)
```

```python
corr_starwars = corr_starwars.join(ratings['rating_vote_number'])
corr_starwars.head()
```

**NOTE:** Films with a small number of votes can mislead us. For example, the fact that 1-2 people gave a film 5 does not mean the same thing as the fact that 100 people gave it 5. For this reason, it would be much more accurate to decide by looking at the average ratings of films with more than 100 votes.

# And the Result

```python
final = corr_starwars[corr_starwars['rating_vote_number']>100].sort_values('Correlation',ascending=False).head()
final
```

```python
final.index[1]
```

```python
print("Final Recommodition: You can also watch the ************", final.index[1], "****************")
```

**FINAL DECISION / OUR MODEL'S RECOMMODITION:** The correlation of Star Wars with "Empire Strikes Back, The (1980)" is 0.74, so it would make sense to recommend **"Empire Strikes Back, The (1980)"** to people **who have seen Star Wars**.
